```python
def lower_case(value):
    print('Here is the first value:',value)
    return value.lower()

data['first_name_min']= data['first name'].apply(lower_case)
data.head(10)
    
```

```python
del data['first_name_min']
```

```python
#This is the list of all valid countries 
VALID_COUNTRIES = ['France','India','China','Benin','Madagascar', 'USA']
```

```python
#This function verifies if the country is in the valid list if not the Nan value is replaced
def check_country(country):
    if country not in VALID_COUNTRIES:
        print('- "{}" is not a valid country, we deleted it.'.format(country))
        return np.NaN
    return country
```

```python
#This function splits the number of emails recorded and takes the first 
#of the multiple emails
def first(string): 
    parts = string.split(',')
    first_part = parts[0]
    if len(parts)>=2:
        print('- There are multiple parts of "{}", use only {}.'.format(parts,first_part))
        return first_part
    else:
        return string 
```

```python
#use Regular expressions to match the format of the height 'm' meters
def convert_height(height):
    found = re.search('\d\.\d{2}m',height)
    if found is None:
        print('- "{}" is in the wrong format. It will be ignored.'.format(height))
        return np.NaN
    else:
        value = height[:-1]# we remove the last character, which is 'm'
        return float(value)
    
def fill_height(height, replacement):
    if pd.isnull(height):
        print('Imputation by the mean : {}'.format(replacement))
        return round(replacement,3)
    return height


        
```

```python
data['email'] = data['email'].apply(first)
data['country'] = data['country'].apply(check_country)
```

```python
data.head(10)
```

```python
data['date_of_birth']= pd.to_datetime(data['date_of_birth'],format='%d/%m/%Y',errors='coerce')
```

```python
data['height'] = [convert_height(t) for t in data['height']]
data['height'] = [t if t<3 else np.NaN for t in data['height']]

mean_height = data['height'].mean()
data['height'] = [fill_height(t,mean_height) for t in data['height']]
```
